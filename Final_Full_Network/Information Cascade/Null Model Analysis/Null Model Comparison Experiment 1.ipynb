Calculate the average number of nodes in a cascade

```python
def count_nodes_per_cascade(df):
    lst = []
    for column in df:
        lst.append(df[column].count())
    return lst


def avg_nodes_per_cascade(df):
    count = 0;
    i
    for column in df:
        i=i+1
        count = count + df[column].count()
    return count / i
```

```python
def generate_stats(df):
    counts = np.array(count_nodes_per_cascade(df))
    print("Average Number of Nodes Activated: {}".format(np.average(counts)))
    print("Standard Deviation of Number of Nodes Activated: {}".format(np.std(counts)))
    print("Maximum Time Activated: {}".format(max(df.max())))
```

```python
generate_stats(df_lib)
print("------------------")
generate_stats(df_con)
print("------------------")
generate_stats(df_ndp)
print("------------------")
generate_stats(df_lgbt)
print("------------------")
generate_stats(df_bloc)
```
