## Identifying Important Bridges
Go through each cascade experiment and see if there are any strong,regular, or weak bridges that were impacted during the cascade

```python
strong_df = pd.read_csv('../Community Analysis/Strong_Bridges.csv')
strong_df
```

```python
reg_df = pd.read_csv('../Community Analysis/Regular_Bridges.csv')
```

```python
weak_df = pd.read_csv('../Community Analysis/Weak_Bridges.csv')
```

### Liberal Cascade

```python
lib_df =  pd.read_csv('liberal_Information_Cascade.csv')
lib_df = lib_df.rename(columns={"Unnamed: 0":"Id"})
```

#### Strong Bridges

```python
strong_lib = pd.merge(lib_df, strong_df, on="Id")
```

```python
strong_lib
```

#### Regular Bridges

```python
reg_lib = pd.merge(lib_df, reg_df, on="Id")
reg_lib
```

#### Weak Bridges

```python
weak_lib = pd.merge(lib_df, weak_df, on="Id")
weak_lib
```
