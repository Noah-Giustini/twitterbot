## Inter Community Interaction in Cascades

```python
nodes = pd.read_csv("../Community Analysis/nodes_with_community_connection_info.csv")
```

```python
def find_comm_interaction(df, nodes,name):
    inter_comm = pd.merge(df, nodes, on="Id")
    inter_comm = inter_comm[inter_comm['community'] != name]
    return inter_comm
```

```python
def cascade_node_info(df, nodes,name):
    inter_comm = pd.merge(df, nodes, on="Id")
    return inter_comm
```

```python
lib_com = find_comm_interaction(lib_df, nodes, "Liberal")
lib = cascade_node_info(lib_df, nodes, "")
con = cascade_node_info(con_df, nodes, "")
bloc = cascade_node_info(bloc_df, nodes, "")
lgbt = cascade_node_info(lgbt_df, nodes, "")
ndp = cascade_node_info(ndp_df, nodes, "")
```
