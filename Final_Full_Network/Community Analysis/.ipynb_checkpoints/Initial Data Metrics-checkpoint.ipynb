```python
def bins_labels(bins, **kwargs):
    bin_w = (max(bins) - min(bins)) / (len(bins) - 1)
    plt.xticks(np.arange(min(bins)+bin_w/2, max(bins), bin_w), bins, **kwargs)
    plt.xlim(bins[0], bins[-1])
```

```python
id_polit = communities[['id','political affiliation']]
id_polit.tail()
```

```python
id_polit.columns
```

```python
edge_list = pd.read_csv('../networkx_df.csv', names=['Source','Target', 'Metadata'])
```

```python
edge_list.head()
```

```python
source_join = pd.merge(id_polit, edge_list, left_on="id", right_on="Source")
```

```python
source_join.head()
```

```python
source_join = source_join.drop('id' ,axis = 1)
```

```python
source_join.head()
```

```python
source_join = source_join.rename(columns={"political affiliation": "source_politic"})
```

```python
source_join.head()
```

```python
target_join = pd.merge(id_polit, source_join, left_on="id", right_on="Target")
```

```python
target_join.head()
```

```python
target_join = target_join.rename(columns={"political affiliation": "target_politic"})
```

```python
target_join.head()
```

```python
target_join = target_join.drop('id' ,axis = 1)
```

```python
target_join.head()
```

```python
subset = lambda a : a[0:len(a)-1]
```

```python
subset('bcde')
```

```python
target_join['Metadata'] = target_join['Metadata'].apply(subset)
```

```python
target_join.head()
```

```python
addattr = lambda a : ',"target_politics": "' + a +'"'
```

```python
addsrc = lambda a : ',"source_politics": "' + a +'"}'
```

```python
addsrc("bcd")
```

```python
target_join.fillna('', inplace=True)
```

```python
target_join.head()
```

```python
target_join['target_politic'] = target_join['target_politic'].apply(addattr)
```

```python
target_join['source_politic'] = target_join['source_politic'].apply(addsrc)
```

```python
target_join.head()
```

```python
target_join['Metadata'] = target_join['Metadata'] + target_join['target_politic'] + target_join['source_politic']
```

```python
target_join.head()
```

```python
target_join.set_index('Source')
```

```python
toReturn = target_join[['Source', 'Target', 'Metadata']]
```

```python
toReturn.head()
```

```python
toReturn.set_index('Source', inplace=True)
```

```python
toReturn.head()
```

```python
toReturn.to_csv("edges_with_affiliations.csv")
```
