```python
import pandas as pd
```

```python
nodes_df = pd.read_csv("Network_export_nodes_betweenness_undirected_with_communities_and_degs.csv")
```

```python
nodes_df.head()
```

```python
edges_df = pd.read_csv("../../Network_export_edges.csv")
```

```python
edges_df = edges_df[["Source","Target","Weight"]]
```

```python
edges_df.head()
```

Join the sources with the id to get the source id betweenness centrality

```python
source_info = pd.merge(
    nodes_df,
    edges_df,
    how="inner",
    on=None,
    left_on="Id",
    right_on="Source",
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)
```

```python
source_info.head()
```

```python
full_info = source_info = pd.merge(
    nodes_df,
    source_info,
    how="inner",
    on=None,
    left_on="Id",
    right_on="Target",
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_Target", "_Source"),
    copy=True,
    indicator=False,
    validate=None,
)
```

```python
full_info.head()
```

## Edges Between Communities

Select communities where communityid_Target, and communityid_Source 

```python
edges_between_communities = full_info[full_info['communityid_Target'] != full_info['communityid_Source']]
```

```python
a =edges_between_communities[['Id_Target','Id_Source', 'communityid_Target', 'communityid_Source', 'betweenness_centrality_Source', 'betweenness_centrality_Target']]
```

```python
a.to_csv("Betweeenness_communities.csv")
```
