```python
import pandas as pd 
import networkx as nx
from networkx.algorithms.community import greedy_modularity_communities
```

```python
# Using readlines()
file1 = open('activist_network_df.csv', 'r')
Lines = file1.readlines()

f1 = open("activists_networkx_format_network.txt", "w")
 

# Strips the newline character
for line in Lines:
    columns = line.split(',')
    f1.write(columns[0] + ","+ columns[1] + "," + '{ "weight": '+ columns[2].strip() + '}\n')
```

```python
G = nx.read_edgelist('activists_networkx_format_network.txt', comments='#',
                     create_using=nx.Graph(), 
                     delimiter=',', 
                     nodetype=int, 
                     encoding='utf-8')
```

```python
c = list(greedy_modularity_communities(G, weight="weight", resolution=0.01))
```

```python
len(c)
```

```python
import networkx.algorithms.community as nx_comm

nx_comm.modularity(G, c, weight="weight")
```
