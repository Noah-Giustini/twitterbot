```python
# Using readlines()
file1 = open('network_df.csv', 'r')
Lines = file1.readlines()

f1 = open("networkx_format_network.txt", "a")
 

# Strips the newline character
for line in Lines:
    columns = line.split(',')
    f1.write(columns[0] + ","+ columns[1] + "," + "{"+ columns[2].strip() + "}\n")
    
f1.close()
    
```
