```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
activist_tweets = pd.read_csv("tweets.csv")
```

```python
activist_tweets.head()
```

```python
activist_retweet_edges = pd.read_csv("activist_tweets_retweet_edges.csv")
```

# Take out the extraneous tweets

```python
import datetime
  
# date in yyyy/mm/dd format
d1 = datetime.datetime(2021, 10, 13)
```

```python
activist_tweets['date'] = pd.to_datetime(activist_tweets['date']).dt.strftime('%Y-%m-%d')
```

```python
activist_tweets.head()
```

```python
#activist_tweets['date'] = activist_tweets['date'].apply(lambda x:x.timestamp())
```

```python
activist_tweets.drop(activist_tweets['date'] > d1.timestamp(), axis = 0, inplace=True)
```

```python
activist_tweets.head()
```

```python
activist_tweets = activist_tweets[activist_tweets['date']<= '2021-10-13']
```

```python
activist_tweets.head()
```

```python
activist_tweets.describe()
```

```python
# Run a join with the edges 
```

```python
activist_retweet_edges.head()
```

```python
joined = pd.merge(activist_retweet_edges, activist_tweets, left_on='tweetid', right_on='id')
```

```python
joined.describe()
```

```python
activist_retweet_edges.describe()
```

```python
# Make userid connect to retweeter id
activist_user_to_retweet_edges = joined[['retweeterid','UserID']]
```

```python
activist_user_to_retweet_edges.to_csv("activist_network.csv")
```
