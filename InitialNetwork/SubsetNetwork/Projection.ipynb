# Projection of Initial Network

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
G = nx.read_edgelist('tweet_retweet_edge.csv', comments='#',
                     create_using=nx.Graph(), 
                     delimiter=',', 
                     nodetype=int, 
                     encoding='utf-8')
```

```python
nx.is_bipartite(G)
```

Load in user nodes

```python
retweeters = pd.read_csv('retweeterNodes.csv')
```

```python
list(retweeters['Id'])
```

```python
Projection =  nx.bipartite.projected_graph(G,list(retweeters['Id']))
```

```python
nx.write_edgelist(Projection, "full_projection.csv", data=False)
```
