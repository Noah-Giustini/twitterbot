# Answering Research Question 2

Can we identify activist organizations as non-partisan or partisan based on their interactions?

```python
import pandas as pd
```

```python
nodes = pd.read_csv("nodes_with_community_connection_info.csv")
```

```python
#get only activist organizations from the nodes
ao = nodes[nodes['political affiliation'] =='Activist Organization']
```

Reminder: 

Target Comm Interaction Count means that you have people from many communities who retweet you
Source Comm Interaction Count means that you retweet people from many communities

## Partisan Organizations
These are described as organizations who only have followers from their own community retweeting them

```python
#get only organizations who have retweeters in the same community 
partisan_ao = ao[ao['Target Comm Interaction Count'] == 0]
```

```python
partisan_ao[['Target Comm Interaction Count', 'id', 'community','political affiliation']]
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(10, 8))
#Get a count of how many organizations in each community are partisan
data = partisan_ao["community"].to_numpy()

N, bins, patches = ax.hist(data, edgecolor='white', bins=10, linewidth=1)

#set colour of known communities
patches[0].set_facecolor('orange')
patches[1].set_facecolor('magenta')
patches[2].set_facecolor('red')

for i in range(3, len(patches)):
    patches[i].set_facecolor('grey')

plt.xlim([0, 9])
plt.xticks(np.arange(0, 10, 1.0))
plt.xticks(rotation=270)
plt.xlabel("Community")
plt.ylabel("Number of Activist Groups")
plt.grid(axis='y')
plt.title(label="Number of Activist Groups with Only Same Community Retweeters per Community")

plt.show()
```
